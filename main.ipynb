```python
complication = data[data['complication']==0]
no_complication = data[data['complication']==1]

print(f'Complication: {len(complication)}')
print(f'No Complication: {len(no_complication)}')
print(f'Percentage of no complication: {len(no_complication)*100/len(complication)}')
```

```python
corrmat = data.corr()
fig = plt.figure(figsize=(12,9))

sns.heatmap(corrmat, vmax = 0.8, square=True)
plt.show()
```

```python
X = data.drop(['complication'], axis=1)
Y = data['complication']
X_data = X.values
Y_data = Y.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data,
                                                    test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,Y_train)
y_pred = rfc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(Y_test,y_pred)
print(conf_matrix)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,matthews_corrcoef
print("The Model used is Random Forest")

acc = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {acc}")

prec = precision_score(Y_test, y_pred)
print(f"Precision: {prec}")

rec = recall_score(Y_test, y_pred)
print(f"Recall: {rec}")

f1 = f1_score(Y_test, y_pred)
print(f"F1: {f1}")

mcc = matthews_corrcoef(Y_test, y_pred)
print(f"Matthews Correlation: {mcc}")
```
